Hertz model:

$$a = \sqrt[3]{2.67 \times 10^4 \frac{\mathrm{nm^3}}{\mu \mathrm{N}}.F}$$

JKR model:

$$a^3 = \frac{3}{4}\frac{R^*}{E^*} \left(F + 3 \pi W R^* +\sqrt{6\pi WR^*F + (3\pi WR^*)^2} \right) $$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'serif'
R = 2 #micro meter
gamma = 0.05 # N/m
k = 0.471

def hertz(F):
  return (2.67 * (10**4) *F)**(1/3)

def JKR(F):
  a3 = 0.75 * R/gamma * (F + k + np.sqrt(2*k*F + (k)**2))
  return a3**(1/3) * 10

x1 = np.arange(0, 5, 0.01)
x2 = np.arange(-1, 5, 0.01)
y1 = hertz(x1)
y2 = JKR(x2)

plt.vlines(x=0, ymin=0, ymax=70, color='black', ls='--')
plt.plot(x1, y1, label='Hertz')
plt.plot(x2, y2, label='JKR')
plt.xlabel('Load ($\mu$N)', fontfamily='serif')
plt.ylabel('Coantact radius (nm)', fontfamily='serif')
plt.legend()
plt.show()
```
