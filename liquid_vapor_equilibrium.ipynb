The partial vapor preassure of the components of an ideal solution of two liquids (a binary solution) are related to the liquid composition through;

$$p_1 = x_1p_s^* \, \text{and} \, p_2=x_2 p_2$$

where $p_i^*$ is the vapor pressure of the liquid labeled $i$ and the mole fractions of the liquid components are $x_i$. For a binary mixture, $x_1 +x_2 = 1$. The total vapor pressure of the mixture is thus

$$p=p_1+p_2 = x_1p_1^*+x_2p_2^* = x_1 p_1^* + (1-x_1)p_2^*$$

ad the paartial vapor pressure are therefore

\begin{align}
y1 &= \frac{p_1}{p}=\frac{x_1 p_1^*}{x_1 p_1^* + (1-x_1)p_2^*} \\
y2 &= \frac{p_2}{p}=\frac{x_2 p_2^*}{x_1 p_1^* + (1-x_1)p_2^*}
\end{align}

For ethanol, $p_1^*=44.5 $ mmHg abd for water, $p_2^* = 23.8$ mmHg (both at 298 K). The above equations can be usde to predict the compositions of the vapor phase of a binary mixture of ethanol and water at different compositions, assuming ideal solution behavior.

```python
import numpy as np
import matplotlib.pyplot as plt

# vapor pressure of pure ethanol (1) and water (2) at 298 K in mmHg
p1_star, p2_star = 44.5, 23.8
x1 = np.linspace(0, 1, 100)
y1 = x1 * p1_star / (x1 * p1_star + (1-x1)*p2_star)

plt.plot(x1, y1, c="green", label="Raoult")
# draw a line at y1 = x1 for reference
plt.plot(x1, x1, c="k")
plt.xlabel("x1")
plt.ylabel("y1")
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.show()
```

$$x_1 = \frac{w_1 / M_1}{w_1 M_1 + (100 - w_1)/M_2}$$

```python
def get_mol_frac(w1, M1, M2):
    """Get the mole fraction of a component of a binary mixture.

    The provided w1 is the mass fraction as a percentage of solution mass;
    M1 and M2 are the molar masses of the component compounds.
    """
    return w1/M1 / (w1/M1 + (100-w1)/M2)

M1, M2 = 46.07, 18.02
x, y = get_mol_frac(wx, M1, M2), get_mol_frac(wy, M1, M2)
plt.plot(x, y, marker='.', lw=1, label='Experiment')
plt.fill_between(x, y, x, alpha=0.2)

# A reference line at y1 = x1
plt.plot([0, 1], [0, 1], c='k', lw=1)
plt.plot(x1, y1, lw=1, label='Roult')
plt.xlabel('x1')
plt.ylabel('y1')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.legend()
plt.show()
```
